## Update the metadata

```python
def add_transcript_to_metadata(video_id, transcript, csv_file="video_metadata.csv"):
    """Add transcript to CSV for specific video ID"""
    try:
        # Read existing data
        rows = []
        with open(csv_file, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            fieldnames = reader.fieldnames
            if 'Transcript' not in fieldnames:
                fieldnames += ['Transcript']
            rows = list(reader)

        # Update matching video
        updated = False
        for row in rows:
            if row['Video ID'] == video_id:
                row['Transcript'] = str(transcript)  # Serialize list to string
                updated = True
                break
        
        if not updated:
            print(f"Video ID {video_id} not found")
            return

        # Write back
        with open(csv_file, 'w', encoding='utf-8', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(rows)

        print(f"Updated transcript for {video_id}")

    except Exception as e:
        print(f"CSV error: {str(e)}")
```

```python
add_transcript_to_metadata(video_id, transcript)
```

```python
import requests
import os

api_key=os.environ.get('api_key1')

def check_ai_training_status(video_id):
    """
    Check the third-party AI training status of a YouTube video.

    :param api_key: Your YouTube Data API key
    :param video_id: The ID of the YouTube video to check
    :return: The AI training status (e.g., 'all', 'none', or specific permissions)
    """
    # Define the endpoint and parameters
    url = "https://youtube.googleapis.com/youtube/v3/videoTrainability"
    params = {
        "id": video_id,
        "key": api_key
    }

    # Make the GET request
    response = requests.get(url, params=params)

    # Handle the response
    if response.status_code == 200:
        data = response.json()
        if "permitted" in data:
            return data["permitted"]
        else:
            return "No information available for this video."
    else:
        return f"Error: {response.status_code}, {response.text}"

VIDEO_ID ='55kmZwUYsPI'

# Call the function and print the result
status = check_ai_training_status(VIDEO_ID)
print(f"AI Training Status for Video {VIDEO_ID}: {status}")
```
